```python
import osmnx as ox
import matplotlib.pyplot as plt
import geopandas as gpd
import pandas as pd

ox.__version__
```

```python
# PROJEÇÃO EM LAGES
neighborhood_info = {
    "Centro": "red",
    "Guadalupe": "blue",
    "Frei Rogério": "purple",
    "São Cristovão": "green"
}
tags = {"building": True}
```

```python
# BAIRROS
#combined_neighborhoods = gpd.GeoDataFrame()

# Retrieve and store the neighborhoods' geometries
for name, color in neighborhood_info.items():
    place_name = f"{name}, Lages, Santa Catarina, Brazil"
    neighborhood = ox.features_from_place(place_name, tags=tags)
    neighborhood["color"] = color  # Add a new 'color' column
    combined_neighborhoods = pd.concat([combined_neighborhoods, neighborhood], ignore_index=True)

fig, ax = plt.subplots(figsize=(15, 15))

for color, subset in combined_neighborhoods.groupby("color"):
    subset.plot(ax=ax, fc="none", ec=color)
ax.set_title("atacados")
ax.axis("off")

#plt.savefig("neighborhoods_plot.png", dpi=300, bbox_inches="tight")

plt.show()
```
