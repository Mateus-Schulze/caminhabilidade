```python
import osmnx as ox
import matplotlib.pyplot as plt

# Define the names of the neighborhoods
neighborhood_names = ["Centro", "Guadalupe"]

# Geocode and project the neighborhoods
neighborhoods = [ox.geocode_to_gdf(f"{name}, Lages, Santa Catarina, Brazil") for name in neighborhood_names]
neighborhoods_proj = [ox.project_gdf(neighborhood) for neighborhood in neighborhoods]

# Plot the neighborhoods without the city background
fig, ax = plt.subplots()
for neighborhood_proj in neighborhoods_proj:
    neighborhood_proj.plot(ax=ax, fc="none", ec="red")

# Set axis off
ax.axis("off")

# Display the plot
plt.show()
```

```python
place_name = "Centro, Lages, Santa Catarina, Brasil"
tags = {"building" : True}
# dist=6500
gdf = ox.features_from_place(place_name, tags)
gdf.shape
gdf_proj = ox.project_gdf(gdf)
fix, ax = ox.plot_footprints(gdf_proj, dpi=300,)
len(gdf)
```
